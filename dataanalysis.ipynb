```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("Merged_COVID_Vaccine_Data.csv")
```

```python
top = df.sort_values('Total Cases',ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='Total Cases',y='State/UTs',hue='Total Cases',palette='rocket',edgecolor='black',linewidth=1.5, data=top)
plt.title('Top 10 States by Total COVID Cases',fontsize=18,fontweight='bold')
plt.grid(alpha=0.3)
plt.show()
```

```python
top = df.nlargest(10,'Deaths')

plt.figure(figsize=(13,7))
sns.barplot(y='Deaths',x='State/UTs',data=top,hue='Deaths',palette='Reds_r',edgecolor='black')
plt.title('Top 10 States by Deaths',fontsize=18,fontweight='bold')
plt.grid(alpha=0.7,linestyle='--',axis="y")
plt.show()

```

```python
top = df.nlargest(10,'Active')
plt.figure(figsize=(14,7))

sns.barplot(data=df,x='State/UTs',y='Active',hue='Active',palette='viridis')

plt.xticks(rotation=90)
plt.title('Active Cases by State',fontsize=18,fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.5)
```

```python
import matplotlib.pyplot as plt


dose1_total = df['Dose1'].sum()
dose2_total = df['Dose 2'].sum()

labels = ['Dose 1', 'Dose 2']
sizes = [dose1_total, dose2_total]

plt.figure(figsize=(8,8))

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True
)

plt.title('Comparison of Total Dose 1 and Dose 2 Vaccinations',
          fontsize=16,
          fontweight='bold')

plt.show()
```

```python
labels = ['Dose1 15-18', 'Dose2 15-18',
          'Dose1 12-14', 'Dose2 12-14']

values = [
    df['Dose 1 15-18'].sum(),
    df['Dose 2 15-18'].sum(),
    df['Dose 1 12-14'].sum(),
    df['Dose 2 12-14'].sum()
]

plt.figure(figsize=(8,8))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Vaccination Distribution')
```
